```python
import numpy as np
import sympy as sym
from matplotlib import pyplot as plt

from pylab import rcParams 
rcParams['figure.figsize'] = 14,8 # Увеличила размеры вывода графиков
```

```python
# Уровень 0. Посчитайте значение первой и второй производной функции: $f(x) = x^5 + 4sin(2x) + cos(3x+3)$ в точке x = 1.
def f(x): # Создала функцию
    return x**5+4*sym.sin(2*x)+sym.cos(3*x+3)
def f1(x): # Создала функцию первой производной
    return sym.diff(f(x))
def f2(x): # Создала фунуцию второй производной
    return sym.diff(f1(x))
x = sym.Symbol('x') # обозначила переменную х
print('Значение первой производной при х=1: ', f1(x), ' = ', f1(x).evalf(subs = {x: 1}))
print('Значение второй производной при х=1: ', f2(x), ' = ', f2(x).evalf(subs = {x: 1}))
```

```python
# Уровень 1. Постройте график первой и второй производной функции: $f(x) = sin^5(2x+1)$ на отрезке [-5; 5].
```

```python
def f(x): # Создала функцию 
    return (sym.sin(2*x+1))**5
def f1(x): # Создала функцию первой производной
    return sym.diff(f(x))
def f2(x): # Создала фунуцию второй производной
    return sym.diff(f1(x))

x = sym.Symbol('x') # обозначила переменную х
print('Значение первой производной при х=1: ', f1(x))
print('Значение второй производной при х=1: ', f2(x))

x1 = np.linspace(-5, 5) # задала массив значений Х на отрезке [-5; 5]
y = [f(x).evalf(subs = {x: xx}) for xx in x1] # задала массив значений первичной функции по х на отрезке [-5; 5]
y1 = [f1(x).evalf(subs = {x: xx}) for xx in x1] # задала массив значений первой производной по х на отрезке [-5; 5]
y2 = [f2(x).evalf(subs = {x: xx}) for xx in x1] # задала массив значений второй производной по х на отрезке [-5; 5]

plt.plot(x1, y, label='function') #Построила график первичной функции
plt.plot(x1, y1, label='diff1') #Построила график первой производной
plt.plot(x1, y2, label='diff2') #Построила график второй производной
plt.legend()
plt.show()
```
