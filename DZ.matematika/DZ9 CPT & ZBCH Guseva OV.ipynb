```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline

#увеличим дефолтный размер графиков
from pylab import rcParams
rcParams['figure.figsize'] = 18,8
```

Создайте случайную величину из любого выбранного вами распределения, для разных значений n сгенерируйте 1000 выборок размера n и постройте гистрограммы средних этих выборок.

```python
rv = sts.norm(loc = 2, scale = 0.5) # условия для генерации выборки
sample = rv.rvs(size=1000) # генерация выборки размером  n = 1000  (Random VariateS).
x = np.linspace(0,5,100)
pdf = rv.pdf(x) # формирование плотности распределения
plt.hist(sample, density=True, bins=50, label='sample histogram')
plt.plot(x, pdf, label='theoretical pdf', alpha=50)
plt.legend()
plt.ylabel('fraction of samples, $f(x)$')
plt.xlabel('$x$')
plt.show()
```

```python
# Параметры запуска: размеры выборок и количество повторов генерации
n1 = 2 # Начальный размер выборок
samples_count = 1000 # Кол-во повторов генераций
a1 = np.array([]) # Массивы для сохранения средних по выборкам

for number in range(samples_count):
    sample1 = rv.rvs(size=n1) # На каждом цикле генерируем выборки разных размеров
    a1 = np.append(a1, sample1.mean()) # Считаем среднее по каждой выборке и добавляем в массив средних
    n1 = n1 + 1
```

```python
# Строим графики
label_1 = 'sample of ' + str(n1-2) + ' histogram'
plt.hist(a1, density=True, bins=150, label=label_1)

plt.legend()
plt.ylabel('samples means')
plt.xlabel('$x$')
```
