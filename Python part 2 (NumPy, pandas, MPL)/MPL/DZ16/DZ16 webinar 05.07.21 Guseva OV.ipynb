```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#увеличим дефолтный размер графиков
from pylab import rcParams
rcParams['figure.figsize'] = 18,8
```

```python
df = pd.read_csv('attacks.csv', encoding='latin-1')
#df.info()
```

```python
#Из статистики удаляю все строки с пустым годом или те, где вместо года указана непонятная информация
#или информация за прошлые века
df = df[df['Year'] > 1900]
df['Year'] = df.Year.astype('Int64')
#df.info()
```

```python
#группирую статистику по странам с подсчетом кол-ва случаев нападений акул и сохраняю информацию по ТОП10 странам
top10_country = df[['Country', 'Case Number']].groupby('Country').count().sort_values(by='Case Number', ascending=False).head(10)
top10_country.plot.barh(title='ТОП-10 Стран по статистике нападения акул на человека.')
print(top10_country)

print('Визуализация проводилась на основе статистических данных по нападению акул на человека в разных странах начиная с 1901 года.\
График 1 отображает ТОП10 стран, в которых наблюдалось больше все случаев нападения акул за все года с 1901 по настоящее время.\
Как видно на графике, первые 3 места занимают такие страны как USA, AUSTRALIA и SOUTH AFRICA.')
plt.show()
```

```python
#Вывожу общую суммарную динамику по нападениям акул на человека суммарно по всем странам в разбивке по годам
dynamics_by_years = df[['Year', 'Case Number']].groupby('Year').count()
dynamics_by_years.plot(title='Общая динамика нападения акул на человека с 1901 года по настоящее время.')

print('Суммарно по всем странам наблюдается рост кол-ва нападений акул на человека.\
Это может быть связано с тем, что в связи с развитием технологических возможностей и доступностью покупки нужного оборудования \
все большее кол-во людей отплывает достаточно далеко от береговой линии для самостоятельного изучения океанических вод и \
профессональной рыбной ловли. На увеличение статистики могло повлиять развитие и рост популярности водных видов спорта типа серфинга или парусных видов спорта.\
Дополнительно, на рост статистики также мог повлиять фактор развития компьютерных технологий. Если ранее и вели статистику и какие либо записи в рукописном виде,\
что нужно было собирать со всех возможных районов и передавать в одно место сбора данных,\
то с развитием компьютерных технолгий такие случаи стали фиксировать и передавать в электронном виде, что увеличило и скорость и качество и своевременность передачи данных. \
Интересно наблюдать общий всплеск нападений в 1960 годах, который также наблюдался и в ТОП странах по отдельности.\
Возможно это происходило с увеличением численности акул. Далее статистика пошла на спад в связи с тем, что стала сильно развиваться охота на акул и защита прибрежных вод.')
plt.show()
```

```python
# Вывожу динамику нападений акул на человека по ТОП3-странам - 'USA', 'AUSTRALIA', 'SOUTH AFRICA'.
# Для этого я я объединяю 2 таблицы: первоначальную и top10_country, удалила ненужные столбцы и создала сводную таблицу
# со столбцами с количеством нападений по каждой стране ТОП10.
# Можно вывести информацию и по всем 10 странам ТОП10, но для наглядности возьмем 3 первые страны с максимальным кол-вом,
# т.к. статистика по остальным странам намного меньше, чем у ТОП3.

top10_country_years = df.merge(top10_country, how='inner', left_on = 'Country', right_on = 'Country')
top10_country_years = top10_country_years.drop(['Case Number_x', 'Location', 'pdf', 'href formula', 'href', 'Case Number.1', 'Case Number.2', 'Unnamed: 22', 'Unnamed: 23'], axis = 1)
tcy = top10_country_years.pivot_table(index = 'Year',
                                      columns = 'Country',
                                      values = 'original order',
                                      aggfunc = 'count').reset_index(level = ['Year'])
tcy.plot(x = 'Year', y = ['USA', 'AUSTRALIA', 'SOUTH AFRICA'], kind='area', stacked=False,
         title='Динамика нападения акул на человека с 1901 года по настоящее время в разбивке по странам ТОП-3.')

print('Для сравнительного анализа кол-ва случаев нападения акул в динамике лет рассмотрим 3 топовых страны.\
В Южной Африке всплески нападения акул наблюдались в 1960гг - 1980 гг - 1990 годах, но после 2000 года рост не наблюдается.\
 В Австралии всплески нападения акул наблюдались в 1930-1940 гг, 1960 гг и начиная с 2000 года у них начался в среднем постоянный рост случаев нападения.\
В Штатах Америки всплески нападения акул наблюдались в 1960 годах, а где то после 1970 года и по настоящее время у них наблюдается в среднем постоянный рост.\
 Помимо причин, перечисленных в общей динамике роста статистики нападений, это также может быть связано с увеличением численности населения в Штатах, например, \
по причине увеличения кол-ва переселенцев из других стран и с увеличением туристического потока в страну  в связи с открытием границ СССР.\
В Австралии численность населения с годами тоже, конечно, росла, но не настолько, как в Штатах.\
В свою очередь в Южной Африке же наоборот, отсутствие технологий в стране могло повлиять на некое занижение статистики по случаям нападения акул.\
 Плюсом, в Южную Африку никто не ездит для занятий водными видами спорта, и туристические потоки сюда не особо интенсивные в том числе, поэтому рост нападений здесь особо не наблюдается.\
 В целом, на рост общей динамики нападений акул на человека сильно влияет динамика роста в США и Австралии.')
plt.show()
```

```python
#Посмотрим статистику случаев нападения акул с точки зрения фатального окончания случая. Закончился ли случай смертью или нет
#по каждой из стран ТОП10.
#Для этого необходимо проверить и выровнить значения столбца 'Fatal (Y/N)'.
#все пустые, неизвестные и непонятные текстовые значения я заполнила 'UN'.
#top10_country_years['Fatal (Y/N)'].unique()

top10_country_years['Fatal (Y/N)'].fillna('UN', inplace = True)
top10_country_years['Fatal (Y/N)'].replace([' N', 'N '], 'N', inplace = True)
top10_country_years['Fatal (Y/N)'].replace(['2017', 'UNKNOWN'], 'UN', inplace = True)

#Создаю сводную таблицу по сводной статистике ТОП10 стран с разбивкой по значеням столбца 'Fatal (Y/N)'.
top10_country_fatal = top10_country_years.pivot_table(index = 'Country',
                                      columns = ['Fatal (Y/N)'],
                                      values = 'original order',
                                      aggfunc = 'count',
                                      fill_value = 0).sort_values(by='Y', ascending=False)
top10_country_fatal.plot.bar(rot = 0, title='Общая статистика нападения акул на человека с 1901 года по настоящее время ТОП10 стран с разбивкой по условию окончания каждого случая')

print('Как видно из статистики, что несмотря на то, что США очень сильно лидирует в целом по общему суммарному количеству \
случаев нападения акул на людей, тем не менее у Австралии количество случаев, смертельно закончившихся явно больше, чем у США. \
Хотя, у США смертельно закончившиеся случаи могут быть скрыты под данными, помеченными как UNKNOWN, которых у США больше, чем у всех остальных стран. \
В целом, интересно наблюдать, что, общая статистика у каких то стран может быть высокой, но если её рассматривать с точки зрения \
смертельно закончившихся случаев, то страна может быть по статистике на одном из последних мест. \
В данном случае необходимо дополнительное исследование причин, по которым происходят нападения акул, и действительно ли \
все нападения акул на людей можно считать действительно за нападения.')

plt.show()
```

```python
# Интересно посмотреть стандартные статистические данные примерно за последние 30 лет по странам ТОП-3.
tcy2 = top10_country_years[top10_country_years.Year>1990].pivot_table(index = 'Year',
                                      columns = 'Country',
                                      values = 'original order',
                                      aggfunc = 'count').reset_index(level = ['Year'])

tcy2.plot(kind='box', y = ['USA', 'AUSTRALIA', 'SOUTH AFRICA'], stacked=False,
         title='Max, Min, среднее значение и медиана по статистике нападения акул на человека за последние 30 лет по странам ТОП-3.')

print('Интересно посмотреть данные примерно за последние 30 лет по странам ТОП-3 \
с учетом предыдущих наблюдений, что в среднем начиная с 1980-1990 гг начался рост случаев нападений. \
Если выводить данный график за весь период с 1901 года, то среднее кол-во нападений в год будет занижено за исключением тех стран, \
у которых кол-во случаев нападений в год не растет слишком сильно.')

tcy.plot(kind='box', y = ['USA', 'AUSTRALIA', 'SOUTH AFRICA'], stacked=False,
         title='Max, Min, среднее значение и медиана по статистике нападения акул на человека c 1901гг по настоящее время по странам ТОП-3.')

plt.show()
```
