```python
print('Статистика по температуре: ')
print('--------------------------------------------------------------------------------------')
print('Cредняя ректальная температура лошади составляет ', round(df.rectal_temp.mean(),3), '. Отклонение от нормы: ', round((df.rectal_temp.mean() - 37.8), 3))
print('Медиана по значениям температуры - ', df.rectal_temp.median(), '. Отклонение от нормы: ', round((df.rectal_temp.median() - 37.8), 3))
print('Температура наиболее часто повторяющаяся - ', df.rectal_temp.mode()[0], '. Она наблюдалась в ', round(len(df.rectal_temp[df.rectal_temp == df.rectal_temp.mode()[0]])/len(df.rectal_temp)*100, 2), ' % случаев.')
print('Максимальное значение температуры - ', df.rectal_temp.max())
print('Минимальное значение температуры - ', df.rectal_temp.min())
print('Размах измеренных температур: ', round(df.rectal_temp.max() - df.rectal_temp.min(), 2))
print('СКО - ', df.rectal_temp.std())
print('Дисперсия - ', df.rectal_temp.var())
q1 = df.rectal_temp.quantile(0.25)
q3 = df.rectal_temp.quantile(0.75)
iqr = q3 - q1
lb = q1 - (1.5 * iqr) 
ub = q3 + (1.5 * iqr)
df2_outliers = df[df.rectal_temp.between(lb, ub, inclusive=True)]
print('Q1 - ', q1, '; Q3 - ', q3, '; IQR: ', round(iqr, 2))
df2_outliers
df3 = pd.concat([df, df2_outliers]).drop_duplicates(keep=False).sort_values('rectal_temp')
df3['rectal_temp'] = df3['rectal_temp'].fillna(0)
print('Кол-во выбросов из основного скопления данных по темп составило: ', len(df3), ', из которых по причине того, что температура не была указана - ', len(df3) - len(df3[df3.rectal_temp != 0]), '. Т.к. медиана, среднее значение и мода практически совпадают, а Q3 и Q1 также не далеки от них, то все пропуски можно заполнить значением Моды.')
print( 'Кол-во актуальных выбросов из основного скопления данных по темп составило:', len(df3[df3.rectal_temp != 0]), '. Они несут допустимые значения и их можно учитывать в рассчетах.')
print('Выбросы без неуказанных значений:')
df3[df3.rectal_temp != 0]

```
