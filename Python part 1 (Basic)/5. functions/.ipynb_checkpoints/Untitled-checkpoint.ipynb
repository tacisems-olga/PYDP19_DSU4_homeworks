### Домашнее задание к лекции "Функции"
Вам нужно помочь секретарю автоматизировать работу. Для этого нужно написать программу, которая будет на основе хранимых данных исполнять пользовательские команды.
Общие требования к программе:
- код должен быть грамотно декомпозирован (каждая функция отвечает за свою конкретную задачу, дублирующийся функционал переиспользуется, а его код не повторяется);
- в коде отсутствуют глобальные переменные (за исключением **documents** и **directories**);
- пользовательский ввод обрабатывается в цикле **while** до тех пор, пока пользователь явно не завершит программу (вводом команды "**q**").

```python
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'},
 {'type': 'insurance2', 'number': '10006', 'name': 'Ольга Гусева'}   
]
directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': ['10006'],
 '4': []
}
command = {
    'p':'Определение владельца документа по его номеру',
    's':'Определение полки хранения документа по его номеру',
    'l':'Вывод полной информации по всем документам',
    'ads':'Добавление новой полки',
    'ds':'Удаление существующей полки из данных (только если она пустая)',
    'ad':'Добавление нового документа в данные',
    'd':'Удаление документа из данных',
    'm':'Перемещение документа с полки на полку',
    'q':'Завершение программы'
}

def p(table_doc):
    """Функция Определения владельца документа по его номеру"""
    doc = input('Введите номер документа: ')
    count = 0
    for i in table_doc:
        if doc == i['number']:
            print('По введенному номеру документа был определен его владелец: ', i['name'])
            count +=1
    if count == 0: print('Документ не найден в базе')
        
def s(table_dir):
    """Функция Определения полки хранения документа по его номеру"""
    doc = input('Введите номер документа: ')
    count = 0
    for i in table_dir:
        if doc in table_dir[i]:
            print('Документ хранится на полке: ', i)
            count +=1
    if count == 0: print('Документ не найден в базе')
        
def l(table_doc,table_dir):
    """Функция Вывода полной информации по всем документам"""
    for i in table_doc:    
        print('№:', i['number'], ', тип: ', i['type'], ', владелец: ', i['name'], ', полки хранения:', end = ' ')
        count = 0        
        for i2 in table_dir:
            if i['number'] in table_dir[i2]:
                count +=1
                if count == 1: print(i2, end = ' ')
                else: print(', ', i2, end = ' ')
        print()    
    
def ads(table_dir):
    """Фунция добавления новой полки"""
    global directories
    sh = input('Введите номер полки: ')
    if sh in table_dir:
        print('Такая полка уже существует. Текущий перечень полок: ', end = ' ')
    else:
        table_dir[sh] = []
        print('Полка добавлена. Текущий перечень полок: ', end = ' ')
    pr(table_dir)
    
def pr(table_dir):
    """Вспомогательная функция вывода текущих номеров полок в одну строку с разделителем ',' между ними """
    count = 0
    for i in table_dir:
        count +=1
        if count < len(table_dir): print(i, ', ', end = ' ')
        else: print(i, '.')
    
def ds(table_dir):
    """Функция - Удаление существующей полки из данных (только если она пустая)"""
    global directories
    sh = input('Введите номер полки: ')
    if sh not in table_dir:
        print('Такой полки не существует. Текущий перечень полок: ', end = ' ')
    else:
        if table_dir[sh] != []:
            print('На полке есть документы, удалите или переместите их перед удалением полки.','Текущий перечень полок: ', sep = "\n", end = ' ')
        else:
            del table_dir[sh]
            print('Полка удалена. Текущий перечень полок: ', end = ' ')
    pr(table_dir)
            
def ad(table_doc,table_dir):
    """Функция - Добавление нового документа в данные"""
    global documents
    global directories
    doc = {
        'type': input('Введите тип документа: '),
        'number': input('Введите номер документа: '),
        'name': input('Введите владельца документа: ')
    }
    sh = input('Введите полку для хранения: ')
    if sh not in table_dir:
        print('Такой полки не существует. Добавьте полку командой as.', 'Текущий список документов:', sep = "\n")
    else:
        table_doc.append(doc)
        table_dir[sh].append(doc['number'])
        print('Документ добавлен. Текущий список документов:')
    l(table_doc,table_dir)  

def d(table_doc,table_dir):
    """Функция - Удаление документа из данных"""    
    global documents
    global directories
    doc = input('Введите номер документа: ')
    count = 0
    temp = list()
    for i in table_doc:
        if doc != i['number']:
            temp.append(i)
        else: count +=1
    if count == 0:
        print('Документ не найден в базе.', 'Текущий список документов:', sep = "\n")        
    else:
        table_doc.clear()
        table_doc +=temp
        #table_doc = list(temp) - не могу понять, почему в данном случае никакое копирование списка temp не работает?
        ddsh(table_dir, doc)
        print('Документ удален.', 'Текущий список документов:', sep = "\n")        
    l(table_doc,table_dir)
        
def ddsh(table_dir,doc):
    """Вспомогательная функция удаления заданного номера документа из реестра полок"""
    global directories
    for i2 in table_dir:
        temp1 = list()
        for i3 in table_dir[i2]:
            if doc != i3:
                temp1.append(i3)
        table_dir[i2] = list(temp1)     
               
def m(table_doc,table_dir):
    """Перемещение документа с полки на полку"""
    doc = input('Введите номер документа: ')    
    с_doc = 0
    for i in table_doc:
        if doc == i['number']: с_doc +=1
    if с_doc == 0:
        print('Документ не найден в базе.', 'Текущий список документов:', sep = "\n")
        l(table_doc,table_dir)
    else:
        sh = input('Введите номер полки: ')
        c_sh = 0
        for i1 in table_dir:
            if sh == i1: c_sh +=1    
        if c_sh == 0:
            print('Такой полки не существует. Текущий перечень полок: ', end = ' ')
            pr(table_dir)
        else:
            ddsh(table_dir,doc)
            table_dir[sh].append(doc)
            print('Документ перемещен.', 'Текущий список документов:', sep = "\n")
            l(table_doc,table_dir)
            
            
print('Вы запустили программу учета документов. Для её завершения нажмите на клавиатуре букву "q".')
a = 0        
while a != 'q':
    a = input('Введите команду: ')
    if a not in command.keys():
        print('Вы пытаетесь запустить несуществую команду')
    else:
        print('Вы запустили команду: ', command[a])
        if a == 'p':
            p(documents)
        if a == 's':
            s(directories)
        if a == 'l':
            l(documents, directories)
            #print(documents)
            #print(directories)
        if a == 'ads':
            ads(directories)
        if a == 'ds':
            ds(directories)
        if a == 'ad':
            ad(documents, directories)
        if a == 'd':
            d(documents, directories)
        if a == 'm':
            m(documents, directories)




```
